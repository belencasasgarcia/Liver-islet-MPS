## Calculate hepatic clearance rate from single-liver culture

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Define parameter for operating conditions in the MPS
VHepaRG = 3*1e-4 # Volume of HepaRG cells(L)
Vm = 3.4*1e-6 # Volume co-culture medium in each organ compartment (L)

Vliver_human = 0.34 

# Define parameters for translation to human in the MPS
```

```python
# Calculate hepatic clearance rate at day 1
# Experimental data
I_0h = np.array([488.9808, 492.8616, 482.4603, 489.4218]) #Insulin at t=0h, day 1, in single-liver culture, four replicates
I_24h = np.array([256.1643, 299.1996, 265.8663, 274.9824]) #Insulin at t=24h, day 1, in single-liver culture, four replicates

print(I_0h.dtype)
I_mean = [I_0h.mean(),I_24h.mean()]
I_SEM = [I_0h.std()/np.sqrt(len(I_0h)),I_24h.std()/np.sqrt(len(I_0h))]
print(I_SEM)
print(len(I_0h))

t = np.array([0, 24])
plt.errorbar(t, I_mean, yerr=I_SEM, fmt="o")
plt.xlabel("time (h)")
plt.ylabel("Insulin concentration (mIU/L)")
plt.show()

# Compute the logarithm of the concentration to calculate the clearance rate
I_mean_log=np.log(I_mean)

# Linear regression model
res = stats.linregress(t, I_mean_log)

plt.plot(t, I_mean_log, 'o', label='Experimental data')
plt.plot(t, res.intercept + res.slope*t, 'r', label='y={:.3f}x+{:.3f}'.format(res.slope,res.intercept))
plt.xlabel("time (h)")
plt.ylabel("Ln Insulin concentration (mIU/L)")
plt.title("Day 1")
plt.legend()


plt.savefig('Insulin_clearance_day1.pdf')


print(res)
print(f"The hepatic clearance rate in the single-liver culture is: {res.slope*(-1):.3f} (1/h)")


```

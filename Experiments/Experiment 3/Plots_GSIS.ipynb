```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
GSISExp3 = pd.read_csv('GSISExp3.csv')
print(GSISExp3)
print(GSISExp3['GSIS'])
GSISExp3['GSIS']=GSISExp3['GSIS']*50 
custom_palette = ["red","green","yellow"]
sns.set_palette(custom_palette)
ax=sns.catplot(x="Glucose", y="GSIS", data=GSISExp3)
ax.set(xlabel='Glycemic level', ylabel='GSIS (fmol/2h)', title='Experiment 3')

ax.savefig("Experiment3.pdf", transparent=True)
```
